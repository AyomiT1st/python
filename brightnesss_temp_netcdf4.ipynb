```python
import xarray as xr
import matplotlib.pyplot as plt

# Open the NetCDF file
ds = xr.open_dataset('C:/Users/taiwoajayi/Downloads/merg_2023081823_4km-pixel.nc4 (1).nc4')

# Access the brightness temperature variable
brightness_temperature = ds['Tb']

# Select a specific time index
time_index = 0
brightness_temperature_at_time = brightness_temperature.isel(time=time_index)

# Create a 2D map plot
plt.figure(figsize=(10, 5))  # Set the figure size as needed
brightness_temperature_at_time.plot.imshow(
    cmap='jet',
    add_colorbar=True,  # Add a colorbar
)

# Customize the plot as needed
plt.title('Brightness Temperature Map')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Set the aspect ratio to be equal
plt.gca().set_aspect('equal', adjustable='box')

# Show the plot
plt.show()

# Close the NetCDF file
ds.close()
```

```python
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs

# Open the NetCDF file
ds = xr.open_dataset('C:/Users/taiwoajayi/Downloads/merg_2023081823_4km-pixel.nc4 (1).nc4')

# Access the brightness temperature variable
brightness_temperature = ds['Tb']

# Select a specific time index
time_index = 0
brightness_temperature_at_time = brightness_temperature.isel(time=time_index)

# Define the extent for the map (West Coast and surrounding ocean)
extent = [-150, -70, 0, 40]  # [min_longitude, max_longitude, min_latitude, max_latitude]

# Create a map plot with the specified extent
plt.figure(figsize=(10, 10))  # Set the figure size
ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_extent(extent, crs=ccrs.PlateCarree())

# Plot the brightness temperature on the map
im = brightness_temperature_at_time.plot.imshow(
    cmap='jet',
    ax=ax,
    transform=ccrs.PlateCarree(),
    add_colorbar=False,
)

# Create a colorbar manually and position it at the base of the map
cbar = plt.colorbar(im, ax=ax, orientation='horizontal', pad=0.05, shrink=0.7)  # Adjust pad and shrink as needed
cbar.set_label('brightness_temperature [K]')

# Customize the map appearance
ax.coastlines()
ax.add_feature(cfeature.RIVERS)
ax.add_feature(cfeature.BORDERS)
ax.add_feature(cfeature.LAND, facecolor='lightgray')
ax.gridlines(draw_labels=True, linewidth=0.5, linestyle='--', color='gray')

# Add title and labels
plt.title('Brightness Temperature Map (West Coast)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Show the plot
plt.show()

# Close the NetCDF file
ds.close()
```
